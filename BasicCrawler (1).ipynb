<h2>Wikipedia Crawler - Marvel Movies</h2>




I am scraping data from a Wikipedia page. My final goal is to extract list of my favorite series of movies of Marvel, I will collect information about the release date, producer and etc..
Let’s learn with doing this project step wise step:

<h3> 1. Import necessary libraries: </h3>

```python
#import the library used to query a website
import urllib
from urllib.request import urlopen
```

```python
#specify the url
wiki = "https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films"
```

```python
#Query the website and return the html to the variable 'page'
page = urlopen(wiki)
```

```python
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page)
```

<h3> 2. Use function “prettify” to look at nested structure of HTML page </h3>

```python
print(soup.prettify())
```

Above, you can see that structure of the HTML tags. This will help you to know about different available tags and how can you play with these to extract information. The Inspect option is also for very helpful for this purpose

<h3> 3. Work with HTML Tags </h3>

<b>A. soup.&lt;tag&gt;:</b> Return content between opening and closing tag including tag.

```python
soup.title
```

<b>B. soup.&lt;tag&gt;.string:</b> Return string within given tag

```python
soup.title.string
```

<b> C. Find all the links within page’s &lt;a&gt; tags:</b> We know that, we can tag a link using tag &lt;a&gt;. So, we should go with option soup.a and it should return the links available in the web page. Let’s do it.

```python
soup.a
```

Above, you can see that, we have only one output. Now to extract all the links within &lt;a&gt;, we will use find_all().

```python
soup.find_all("a")
```

Above, it is showing all links including titles, links and other information.  Now to show only links, we need to iterate over each a tag and then return the link using attribute “href” with get

```python
links = soup.find_all("a")
for link in links:
    print(link.get("href"))
```

If you pass in a regular expression object, Beautiful Soup will filter against
that regular expression using its match() method. 

This code finds all the tags whose names start with the letter "b",
in this case, the 'body' tag, the 'b' tag and the 'br' tage:

```python
import re
for tag in soup.find_all(re.compile("^b")):
    print(tag.name)
```

<h3> 4. Find the right table </h3> 
<br>
As we are seeking a table to extract information about marvel movies, we should identify the right table first. Let’s write the command to extract information within all <b>table</b> tags.

```python
all_tables=soup.find_all('table')
```

Now to identify the right table, we will use attribute “class” of table and use it to filter the right table. In chrome, you can check the class name by right click on the required table of web page –> Inspect element –> Copy the class name OR go through the output of above command find the class name of right table.

```python
right_table=soup.find('table', class_='wikitable plainrowheaders')
```

```python
print(right_table.prettify())
```

<h3> 5. Extract the information to DataFrame and store </h3> <br> Here, we need to iterate through each row (tr) and then assign each element of tr (td) to a variable and append it to a list. Let’s first look at the HTML structure of the table (I am not going to extract information for table heading <th>). 

Above, you can notice that second element of &lt;tr&gt; is within tag &lt;th&gt; not &lt;td&gt; so we need to take care for this. Now to access value of each element, we will use “find(text=True)” option with each element.  (what is th?) Let’s look at the code:

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') #To store second column data
    if len(cells)==4: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame(A,columns=['U.S. release date'])
df['Film']=B
df['Director(s)']=C
df['Screenwriter(s)']=D
df['Producer(s)']=E
df
```

```python
# store df in file (db is also possible)
from pandas import ExcelWriter

writer = ExcelWriter('PythonExport.xlsx')
df.to_excel(writer)
writer.save()
```
